# Python Visualisation Scripts (Preliminary)

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

# Import data from data folder
trips_data = pd.read_csv("../data/trips_data.csv")
```

## Correlation heatmaps between variables

```python
plt.subplots(figsize=(7,7))
seaborn.heatmap(trips_data.corr(), cmap='vlag') 
```

From the correlation heatmap, variables with the highest correlations (both positive and negative) are `trip_id` and `station_id`s. However, since the distriubtion of ids do not give us much information, the correlation heatmap does not tell us much information about the data

## Checking Distributions
1. boxplot for duration minutes
2. pie charts for distribution of categorical variables

```python
def boxplot_for_duration_minutes(value = 12000):
  trips_data[trips_data['duration_minutes'] <= value][['duration_minutes']].boxplot()
boxplot_for_duration_minutes() # duration data is still very skewed
```

```python
plt.title("Distribution of trips based on parts_of_day")
trips_data['parts_of_day'].value_counts().plot(kind='pie')
```

```python
plt.title("Distribution of trips based on day of the week")
trips_data['day'].value_counts().plot(kind='pie')
```

From the above diagrams we can see that most trips happen during the early afternoon or late afternoon, while the distribution of trips throughout the days of a week are quite even, with a slight preference to more trips on  weekends.
