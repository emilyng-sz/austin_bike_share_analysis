# Python `journey_data` insights

```python
# Import libraries
import pandas as pd

# Import data from data folder
trips_data = pd.read_csv("../data/trips_data.csv")
journey_data = pd.read_csv("../data/journey_data.csv")
same_start_end_journey_data = pd.read_csv("../data/same_start_end_journey_data.csv")
diff_start_end_journey_data = pd.read_csv("../data/diff_start_end_journey_data.csv")
```

Start by inspecting the overall statistics of `trips_data` before comparing with the data of various journey types

```python
trips_data.describe()
```

Since the top 14 journey have the same start and ending station, it may be more insightful to inspect the datasets with the same start and ending station and different start and ending stations respectively.

```python
same_start_end_journey_data[:15]
```

```python
same_start_end_journey_data[:15].describe()
```

From the top 15 `same_start_end_journey_data`,
1. top trips mostly take place on the Weekends 
2. The average duration is relatively high (range of 44.2 to 65.5 minutes) with mean 54.9 minutes compared to the mean of 28.8 minutes for all trips
3. Most common subscriber type is Walk Up
4. The average starting hour is during the afternoon at 2:18pm (14.29hrs)

From these information, we can infer that **casual cyclers are a large proportion of Bikeshare's market**. These users choose to cycle for long periods during the weekend--potentially to visit places around Austin before returning the bike at the same starting point. **A potential market of Bikeshare is therefore tourists**

```python
diff_start_end_journey_data[:15]
```

```python
diff_start_end_journey_data[:15].describe()
```

From the top 15 `diff_start_end_journey_data`,
1. top trips mostly take place on Weekdays (mean is 3.13 which is appoximately Thursday) 
2. The average duration is relatively low (range of 7.7 to 31.3 minutes) with mean 14.5 minutes compared to the mean of 28.8 minutes
4. Most common subscriber type is Local365, followed by Walk Up
5. The average starting hour is around 1:28pm (13.48hrs) but with a larger distribution compared to `journey_data` (1.366 compared to 0.377)

From these information, we can infer that another market of Bikeshare are Local365 i.e. **yearly subscribers according to their bike pass website: https://austin.bcycle.com/passes, who may use the bike to travel short distances frequently**. A potential scenario would be going to work, which makes sense since City Hall / Lavaca & 2nd is a civic district with many offices. **A potential market of Bikeshare is therefore local office workers.**
